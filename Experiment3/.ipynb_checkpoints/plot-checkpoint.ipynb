```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pickle
```

```python
obj_list = ['MT2', 'MX2', 'APFL2']
dataset_list = ['MNIST', 'KMNIST', 'FMNIST']

loss_results_iter = []
loss_results_commun = []
for obj in obj_list:
    loss1_iter = []
    loss1_commun = []
    for dataset in dataset_list:
        file_iter_name = obj + '_' + dataset + '_CDVR_iter_result'
        file_commun_name = obj + '_' + dataset + '_CDVR_commun_result'
        
        result_iter_path = './result/' + file_iter_name + '.txt'
        with open(result_iter_path, "rb") as f:
            loss = pickle.load(f)
            loss1_iter.append(loss)
        
        result_commun_path = './result/' + file_commun_name + '.txt'
        with open(result_commun_path, "rb") as f:
            loss = pickle.load(f)
            loss1_commun.append(loss)
            
    loss_results_iter.append(loss1_iter)
    loss_results_commun.append(loss1_commun)
```

```python
fig, axes = plt.subplots(3, 3, figsize=[16., 12.])

for i, obj in enumerate(obj_list):
    for j, dataset in enumerate(dataset_list):
        axes[i, j].plot(loss_results_commun[i][j][0], color='b', linestyle='-', label='$p_w$=theory')
        axes[i, j].plot(loss_results_commun[i][j][1], color='r', linestyle='-', label='$p_w$=0.7')
        axes[i, j].plot(loss_results_commun[i][j][2], color='g', linestyle='-', label='$p_w$=0.5')
        axes[i, j].plot(loss_results_commun[i][j][3], color='c', linestyle='-', label='$p_w$=0.3')
        axes[i, j].plot(loss_results_commun[i][j][4], color='m', linestyle='-', label='$p_w$=0.1')
        axes[i, j].tick_params(labelsize=18)
        if j == 0:
            axes[i, j].set_ylabel(obj + '\nLoss', fontsize=25)
        if i == 2:
            axes[i, j].set_xlabel('Communication Rounds\n' + dataset, fontsize=25)

fig.legend(labels=['$p_w$=theory', '$p_w$=0.7', '$p_w$=0.5', '$p_w$=0.3', '$p_w$=0.1'], loc='upper center', ncol=5, fontsize=25)

plt.savefig('experiment3_1.png', dpi=400, bbox_inches='tight')

plt.show()
```
